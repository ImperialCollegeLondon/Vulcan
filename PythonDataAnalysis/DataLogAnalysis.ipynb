```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.gofplots import qqplot
```

```python
filename = "three_cols.csv"
```

```python
def check_normality(filename,mean,std):
    df = pd.read_csv(filename)
    mean = df.iloc[0,0]
    std = df.iloc[1,0]
    df = df.iloc[2:,0].copy()
    print("Expected mean: ",mean, " std: ", std)
    print("Program data mean: ",df.mean(), " std: ", df.std())
    plt.hist(df,color='b',bins=100)
    plt.show()
    qqplot(df, line='s')
    plt.show()
```

```python
check_normality()
```
