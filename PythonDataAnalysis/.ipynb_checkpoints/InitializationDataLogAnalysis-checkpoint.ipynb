QQ Plot function to check for normality

```python
def plot_qq(df,col_name):
    plt.figure(figsize=(4,3))
    stats.probplot(df[col_name], dist="norm", plot=plt)
    plt.title(col_name + " Normal Distribution QQ plot")
    plt.show()
```

## Household data

```python
df_h.head()
```

Get all columns that should be normally distributed (i.e. not all zeros)

```python
df_h_norm = df_h.loc[:,(df_h.sum(axis=0) != 0)]
df_h_norm.head()
```

```python
for col_name in df_h_norm.columns:
    plot_qq(df_h_norm,col_name)
```

```python
def check_normality(filename,mean=0,std=0):
    df = pd.read_csv(filename)
    mean = df.iloc[0,0]
    std = df.iloc[1,0]
    df = df.iloc[2:,0].copy()
    print("Expected mean: ",mean, " std: ", std)
    print("Program data mean: ",df.mean(), " std: ", df.std())
    plt.hist(df,color='b',bins=100)
    plt.show()
    qqplot(df, line='s')
    plt.show()
```
